# Processor

```python
def assign_graphcast_processor_weights(graphcast_model, weights):
    """Assigns weights to the GraphCast processor layers.

    Args:
        graphcast_model: The GraphCast model.
        weights: A dictionary containing the pre-trained weights.
    """
    for i in range(0, 14):
        edge_block_index = i * 2 # Edge blocks are at even indices
        node_block_index = i*2 + 1 # Node blocks are at odd indices
        
        edge_mlp = graphcast_model.processor.processor_layers[edge_block_index].edge_mlp.model
        edge_prefix = f'params:mesh_gnn/~_networks_builder/processor_edges_{i}_mesh_mlp/~'
        edge_ln_prefix = f'params:mesh_gnn/~_networks_builder/processor_edges_{i}_mesh_layer_norm'

        edge_mlp[0].weight.data.copy_(torch.tensor(weights[edge_prefix + '/linear_0:w'].T))
        edge_mlp[0].bias.data.copy_(torch.tensor(weights[edge_prefix + '/linear_0:b'].T))
        edge_mlp[2].weight.data.copy_(torch.tensor(weights[edge_prefix + '/linear_1:w'].T))
        edge_mlp[2].bias.data.copy_(torch.tensor(weights[edge_prefix + '/linear_1:b'].T))
        edge_mlp[3].weight.data.copy_(torch.tensor(weights[edge_ln_prefix + ':scale']))
        edge_mlp[3].bias.data.copy_(torch.tensor(weights[edge_ln_prefix + ':offset']))

        node_mlp = graphcast_model.processor.processor_layers[node_block_index].node_mlp.model
        node_prefix = f'params:mesh_gnn/~_networks_builder/processor_nodes_{i}_mesh_nodes_mlp/~'
        node_ln_prefix = f'params:mesh_gnn/~_networks_builder/processor_nodes_{i}_mesh_nodes_layer_norm'

        node_mlp[0].weight.data.copy_(torch.tensor(weights[node_prefix + '/linear_0:w'].T))
        node_mlp[0].bias.data.copy_(torch.tensor(weights[node_prefix + '/linear_0:b'].T))
        node_mlp[2].weight.data.copy_(torch.tensor(weights[node_prefix + '/linear_1:w'].T))
        node_mlp[2].bias.data.copy_(torch.tensor(weights[node_prefix + '/linear_1:b'].T))
        node_mlp[3].weight.data.copy_(torch.tensor(weights[node_ln_prefix + ':scale']))
        node_mlp[3].bias.data.copy_(torch.tensor(weights[node_ln_prefix + ':offset']))
assign_graphcast_processor_weights(graphcast_model, weights)
```
